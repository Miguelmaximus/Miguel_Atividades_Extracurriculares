## PARTE 4 – Polimorfismo e Banco de Dados

- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inserção (INSERT), consulta (SELECT) e exclusão (DELETE) de registros no banco de dados.

**Código:**

```python
# Desenvolva aqui o código da Parte 4
import sqlite3

class AtividadeExtracurricular:
    def __init__(self, nome, descricao, participantes):
        self.nome = nome
        self.__descricao = descricao
        self.participantes = participantes

    def get_descricao(self):
        return self.__descricao

    def set_descricao(self, nova_descricao):
        if len(nova_descricao) > 10:
            self.__descricao = nova_descricao
        else:
            print("Descrição muito curta.")

    def exibir_dados(self):
        print(f"Atividade: {self.nome}")
        print(f"Descrição: {self.__descricao}")
        print("Participantes:", ", ".join(self.participantes))

    def verificar_participantes_minimos(self, minimo):
        return len(self.participantes) >= minimo



class AtividadeCompetitiva(AtividadeExtracurricular):
    def __init__(self, nome, descricao, participantes, nivel_competicao):
        super().__init__(nome, descricao, participantes)
        self.nivel_competicao = nivel_competicao

   
    def exibir_dados(self):
        super().exibir_dados()
        print(f"Nível da Competição: {self.nivel_competicao.upper()}")




class BancoDeDados:
    def __init__(self, nome_banco="atividades.db"):
        self.conexao = sqlite3.connect(nome_banco)
        self.cursor = self.conexao.cursor()
        self.criar_tabela()

    def criar_tabela(self):
        self.cursor.execute("""
        CREATE TABLE IF NOT EXISTS atividades (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            nome TEXT NOT NULL,
            descricao TEXT,
            participantes TEXT,
            tipo TEXT,
            nivel_competicao TEXT
        )
        """)
        self.conexao.commit()

    def inserir_atividade(self, atividade):
        participantes_str = ",".join(atividade.participantes)
        if isinstance(atividade, AtividadeCompetitiva):
            tipo = "competitiva"
            nivel = atividade.nivel_competicao
        else:
            tipo = "comum"
            nivel = None

        self.cursor.execute("""
        INSERT INTO atividades (nome, descricao, participantes, tipo, nivel_competicao)
        VALUES (?, ?, ?, ?, ?)
        """, (atividade.nome, atividade.get_descricao(), participantes_str, tipo, nivel))
        self.conexao.commit()
        print(f"✅ Atividade '{atividade.nome}' inserida no banco.")

    def listar_atividades(self):
        self.cursor.execute("SELECT * FROM atividades")
        atividades = self.cursor.fetchall()
        for a in atividades:
            print(f"[ID {a[0]}] {a[1]} - {a[3]} participantes - Tipo: {a[4]}")

    def excluir_atividade(self, id_atividade):
        self.cursor.execute("DELETE FROM atividades WHERE id = ?", (id_atividade,))
        self.conexao.commit()
        print(f"❌ Atividade ID {id_atividade} removida.")

    def fechar(self):
        self.conexao.close()


if __name__ == "__main__":
    
    comum = AtividadeExtracurricular(
        "Aula de Yoga", "Atividade relaxante de alongamento e respiração.", ["Alice", "Bruno"]
    )
    comp = AtividadeCompetitiva(
        "Maratona de Programação", "Desafio de programação para equipes.", ["Carlos", "Diana"], "nacional"
    )

   
    comum.exibir_dados()
    print()
    comp.exibir_dados()
    print("\n" + "-"*40)

  
    db = BancoDeDados()
    db.inserir_atividade(comum)
    db.inserir_atividade(comp)

    print("\n🔎 Atividades no banco:")
    db.listar_atividades()

   

    db.fechar()
```

## PARTE 5 – Projeto Final e Organização

**Tarefas:**
- Consolidar todas as partes do projeto em um único script organizado.
- Comentar o código explicando cada parte.
- Testar todas as funcionalidades.

**Código:**

```python
# Desenvolva aqui o código da Parte 5
import sqlite3


# Classe base: AtividadeExtracurricular

class AtividadeExtracurricular:
    """
    Representa uma atividade comum, com nome, descrição e lista de participantes.
    """
    def __init__(self, nome, descricao, participantes):
        self.nome = nome
        self.__descricao = descricao  # Atributo privado
        self.participantes = participantes

    def get_descricao(self):
        return self.__descricao

    def set_descricao(self, nova_descricao):
        # Valida que a nova descrição tenha pelo menos 10 caracteres
        if len(nova_descricao) > 10:
            self.__descricao = nova_descricao
        else:
            print("⚠ Descrição muito curta. Deve ter mais de 10 caracteres.")

    def exibir_dados(self):
        # Exibe os dados da atividade
        print(f"📌 Atividade: {self.nome}")
        print(f"📝 Descrição: {self.__descricao}")
        print("👥 Participantes:", ", ".join(self.participantes))

    def verificar_participantes_minimos(self, minimo):
        # Verifica se há participantes suficientes
        return len(self.participantes) >= minimo



# Classe derivada: AtividadeCompetitiva

class AtividadeCompetitiva(AtividadeExtracurricular):
    """
    Representa uma atividade do tipo competitiva com nível de competição.
    """
    def __init__(self, nome, descricao, participantes, nivel_competicao):
        super().__init__(nome, descricao, participantes)
        self.nivel_competicao = nivel_competicao

    def exibir_dados(self):
        # Exibe os dados da atividade competitiva, incluindo o nível
        super().exibir_dados()
        print(f"🏆 Nível da Competição: {self.nivel_competicao.upper()}")



# Classe: BancoDeDados

class BancoDeDados:
    """
    Controla a conexão com o banco de dados SQLite e operações de CRUD.
    """
    def __init__(self, nome_banco="atividades.db"):
        self.conexao = sqlite3.connect(nome_banco)
        self.cursor = self.conexao.cursor()
        self.criar_tabela()

    def criar_tabela(self):
        # Cria a tabela de atividades se ela ainda não existir
        self.cursor.execute("""
        CREATE TABLE IF NOT EXISTS atividades (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            nome TEXT NOT NULL,
            descricao TEXT,
            participantes TEXT,
            tipo TEXT,
            nivel_competicao TEXT
        )
        """)
        self.conexao.commit()

    def inserir_atividade(self, atividade):
        # Insere uma atividade no banco
        participantes_str = ",".join(atividade.participantes)

        if isinstance(atividade, AtividadeCompetitiva):
            tipo = "competitiva"
            nivel = atividade.nivel_competicao
        else:
            tipo = "comum"
            nivel = None

        self.cursor.execute("""
        INSERT INTO atividades (nome, descricao, participantes, tipo, nivel_competicao)
        VALUES (?, ?, ?, ?, ?)
        """, (atividade.nome, atividade.get_descricao(), participantes_str, tipo, nivel))
        self.conexao.commit()
        print(f"✅ Atividade '{atividade.nome}' inserida no banco.")

    def listar_atividades(self):
        # Lista todas as atividades cadastradas
        self.cursor.execute("SELECT * FROM atividades")
        atividades = self.cursor.fetchall()

        if not atividades:
            print("⚠ Nenhuma atividade cadastrada.")
            return

        print("\n📚 Atividades no Banco de Dados:")
        for a in atividades:
            print(f"[ID {a[0]}] {a[1]} - Tipo: {a[4].capitalize()} - Participantes: {a[3]}")

    def excluir_atividade(self, id_atividade):
        # Exclui uma atividade pelo ID
        self.cursor.execute("DELETE FROM atividades WHERE id = ?", (id_atividade,))
        self.conexao.commit()
        print(f"❌ Atividade ID {id_atividade} removida.")

    def fechar(self):
        # Fecha a conexão com o banco
        self.conexao.close()
        print("🔒 Conexão com o banco encerrada.")


# Bloco principal para testes

if __name__ == "__main__":
    print("🚀 Iniciando testes do sistema de atividades extracurriculares...\n")

    # Criando instâncias de atividades
    comum = AtividadeExtracurricular(
        "Aula de Yoga",
        "Atividade relaxante de alongamento e respiração.",
        ["Alice", "Bruno"]
    )

    comp = AtividadeCompetitiva(
        "Maratona de Programação",
        "Desafio de programação para equipes.",
        ["Carlos", "Diana"],
        "nacional"
    )

    # Testando exibição dos dados
    comum.exibir_dados()
    print()
    comp.exibir_dados()

    # Verificação de participantes
    print("\n👥 Verificação de participantes mínimos:")
    print("Yoga tem participantes suficientes?", comum.verificar_participantes_minimos(2))
    print("Maratona tem participantes suficientes?", comp.verificar_participantes_minimos(3))

    # Conectando ao banco e testando inserção/listagem
    print("\n💾 Testando banco de dados...")
    db = BancoDeDados()

    db.inserir_atividade(comum)
    db.inserir_atividade(comp)

    db.listar_atividades()

    # Teste opcional de exclusão
    # db.excluir_atividade(1)

    db.fechar()

    print("\n✅ Todos os testes foram executados.")
```
