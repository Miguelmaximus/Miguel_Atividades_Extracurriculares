```python
# Desenvolva aqui o c√≥digo da Parte 4
import sqlite3


class AtividadeExtracurricular:
    def __init__(self, nome, descricao, participantes):
        self.nome = nome
        self.__descricao = descricao  
        self.participantes = participantes

    
    def get_descricao(self):
        return self.__descricao

    def set_descricao(self, nova_descricao):
        if isinstance(nova_descricao, str) and len(nova_descricao) > 5:
            self.__descricao = nova_descricao
        else:
            print("Descri√ß√£o inv√°lida. Deve ter mais de 5 caracteres.")

    def exibir_dados(self):
        print(f"Atividade: {self.nome}")
        print(f"Descri√ß√£o: {self.__descricao}")
        print("Participantes:")
        for participante in self.participantes:
            print(f" - {participante}")
        print("-" * 40)

    def adicionar_participante(self, nome_participante):
        self.participantes.append(nome_participante)
        print(f"{nome_participante} foi adicionado(a) √† atividade '{self.nome}'.")

    def remover_participante(self, nome_participante):
        if nome_participante in self.participantes:
            self.participantes.remove(nome_participante)
            print(f"{nome_participante} foi removido(a) da atividade '{self.nome}'.")
        else:
            print(f"{nome_participante} n√£o est√° na lista de participantes.")

    def verificar_participacao(self):
        qtd = len(self.participantes)
        if qtd == 0:
            print("N√£o h√° participantes nesta atividade.")
        elif qtd < 3:
            print(f"Atividade '{self.nome}' com poucos participantes: {qtd}.")
        else:
            print(f"Atividade '{self.nome}' com boa participa√ß√£o: {qtd} participantes.")


class AtividadeEsportiva(AtividadeExtracurricular):
    def __init__(self, nome, descricao, participantes, modalidade):
        super().__init__(nome, descricao, participantes)
        self.modalidade = modalidade

    def exibir_dados(self):
        print(f"[ESPORTE] Atividade: {self.nome}")
        print(f"Modalidade: {self.modalidade}")
        print(f"Descri√ß√£o: {self.get_descricao()}")
        print("Participantes:")
        for participante in self.participantes:
            print(f" - {participante}")
        print("-" * 40)



def criar_tabela():
    conn = sqlite3.connect("atividades.db")
    cursor = conn.cursor()
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS atividades (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            nome TEXT NOT NULL,
            descricao TEXT NOT NULL
        )
    """)
    conn.commit()
    conn.close()

def inserir_atividade(nome, descricao):
    conn = sqlite3.connect("atividades.db")
    cursor = conn.cursor()
    cursor.execute("INSERT INTO atividades (nome, descricao) VALUES (?, ?)", (nome, descricao))
    conn.commit()
    conn.close()
    print(f"Atividade '{nome}' inserida no banco de dados.")

def consultar_atividades():
    conn = sqlite3.connect("atividades.db")
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM atividades")
    atividades = cursor.fetchall()
    conn.close()
    print("\nüìã Lista de atividades no banco:")
    for id, nome, descricao in atividades:
        print(f"ID: {id} | Nome: {nome} | Descri√ß√£o: {descricao}")
    print("-" * 40)

def excluir_atividade_por_id(id):
    conn = sqlite3.connect("atividades.db")
    cursor = conn.cursor()
    cursor.execute("DELETE FROM atividades WHERE id = ?", (id,))
    conn.commit()
    conn.close()
    print(f"Atividade com ID {id} exclu√≠da do banco.")



if __name__ == "__main__":
    criar_tabela()

    
    robo = AtividadeExtracurricular("Rob√≥tica", "Projeto com Arduino.", ["Ana", "Lucas"])
    futebol = AtividadeEsportiva("Futebol", "Jogo de sal√£o", ["Carlos", "Jo√£o"], "Futsal")

    robo.exibir_dados()
    robo.verificar_participacao()

    futebol.exibir_dados()
    futebol.adicionar_participante("Pedro")

   
    inserir_atividade(robo.nome, robo.get_descricao())
    inserir_atividade(futebol.nome, futebol.get_descricao())

    
    consultar_atividades()

   
    excluir_atividade_por_id(1)
    consultar_atividades()
```

```python
# Desenvolva aqui o c√≥digo da Parte 5

import sqlite3

# ============================
# PARTE 1, 2 e 3: CLASSE PRINCIPAL E SUBCLASSE
# ============================

class AtividadeExtracurricular:
    def __init__(self, nome, descricao, participantes):
        self.nome = nome
        self.__descricao = descricao  # Atributo privado
        self.participantes = participantes

    def get_descricao(self):
        return self.__descricao

    def set_descricao(self, nova_descricao):
        if isinstance(nova_descricao, str) and len(nova_descricao) > 5:
            self.__descricao = nova_descricao
        else:
            print("Descri√ß√£o inv√°lida. Deve ter mais de 5 caracteres.")

    def exibir_dados(self):
        print(f"Atividade: {self.nome}")
        print(f"Descri√ß√£o: {self.__descricao}")
        print("Participantes:")
        for participante in self.participantes:
            print(f" - {participante}")
        print("-" * 40)

    def adicionar_participante(self, nome_participante):
        self.participantes.append(nome_participante)
        print(f"{nome_participante} foi adicionado(a) √† atividade '{self.nome}'.")

    def remover_participante(self, nome_participante):
        if nome_participante in self.participantes:
            self.participantes.remove(nome_participante)
            print(f"{nome_participante} foi removido(a) da atividade '{self.nome}'.")
        else:
            print(f"{nome_participante} n√£o est√° na lista de participantes.")

    def verificar_participacao(self):
        qtd = len(self.participantes)
        if qtd == 0:
            print("N√£o h√° participantes nesta atividade.")
        elif qtd < 3:
            print(f"Atividade '{self.nome}' com poucos participantes: {qtd}.")
        else:
            print(f"Atividade '{self.nome}' com boa participa√ß√£o: {qtd} participantes.")


class AtividadeEsportiva(AtividadeExtracurricular):
    def __init__(self, nome, descricao, participantes, modalidade):
        super().__init__(nome, descricao, participantes)
        self.modalidade = modalidade

    def exibir_dados(self):
        print(f"[ESPORTE] Atividade: {self.nome}")
        print(f"Modalidade: {self.modalidade}")
        print(f"Descri√ß√£o: {self.get_descricao()}")
        print("Participantes:")
        for participante in self.participantes:
            print(f" - {participante}")
        print("-" * 40)

# ============================
# PARTE 4: BANCO DE DADOS COM SQLITE
# ============================

def criar_tabela():
    conn = sqlite3.connect("atividades.db")
    cursor = conn.cursor()
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS atividades (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            nome TEXT NOT NULL,
            descricao TEXT NOT NULL
        )
    """)
    conn.commit()
    conn.close()

def inserir_atividade(nome, descricao):
    conn = sqlite3.connect("atividades.db")
    cursor = conn.cursor()
    cursor.execute("INSERT INTO atividades (nome, descricao) VALUES (?, ?)", (nome, descricao))
    conn.commit()
    conn.close()
    print(f"Atividade '{nome}' inserida no banco de dados.")

def consultar_atividades():
    conn = sqlite3.connect("atividades.db")
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM atividades")
    atividades = cursor.fetchall()
    conn.close()
    print("\nüìã Lista de atividades no banco:")
    for id, nome, descricao in atividades:
        print(f"ID: {id} | Nome: {nome} | Descri√ß√£o: {descricao}")
    print("-" * 40)

def excluir_atividade_por_id(id):
    conn = sqlite3.connect("atividades.db")
    cursor = conn.cursor()
    cursor.execute("DELETE FROM atividades WHERE id = ?", (id,))
    conn.commit()
    conn.close()
    print(f"Atividade com ID {id} exclu√≠da do banco.")

# ============================
# PARTE 5: MENU INTERATIVO
# ============================

def menu():
    criar_tabela()
    atividades = []

    while True:
        print("\n===== MENU DE ATIVIDADES =====")
        print("1. Cadastrar nova atividade")
        print("2. Cadastrar atividade esportiva")
        print("3. Exibir todas as atividades cadastradas (mem√≥ria)")
        print("4. Inserir atividades no banco de dados")
        print("5. Consultar banco de dados")
        print("6. Excluir atividade do banco de dados")
        print("0. Sair")
        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == "1":
            nome = input("Nome da atividade: ")
            descricao = input("Descri√ß√£o: ")
            participantes = input("Participantes (separados por v√≠rgula): ").split(',')
            atividade = AtividadeExtracurricular(nome, descricao, [p.strip() for p in participantes])
            atividades.append(atividade)
            print("Atividade comum cadastrada com sucesso.")

        elif opcao == "2":
            nome = input("Nome da atividade esportiva: ")
            descricao = input("Descri√ß√£o: ")
            participantes = input("Participantes (separados por v√≠rgula): ").split(',')
            modalidade = input("Modalidade esportiva: ")
            atividade = AtividadeEsportiva(nome, descricao, [p.strip() for p in participantes], modalidade)
            atividades.append(atividade)
            print("Atividade esportiva cadastrada com sucesso.")

        elif opcao == "3":
            if atividades:
                for atividade in atividades:
                    atividade.exibir_dados()
            else:
                print("Nenhuma atividade cadastrada na mem√≥ria.")

        elif opcao == "4":
            for atividade in atividades:
                inserir_atividade(atividade.nome, atividade.get_descricao())

        elif opcao == "5":
            consultar_atividades()

        elif opcao == "6":
            try:
                id_del = int(input("Digite o ID da atividade a ser exclu√≠da: "))
                excluir_atividade_por_id(id_del)
            except ValueError:
                print("ID inv√°lido. Digite um n√∫mero inteiro.")

        elif opcao == "0":
            print("Saindo do sistema. At√© logo!")
            break

        else:
            print("Op√ß√£o inv√°lida. Tente novamente.")

# Executa o menu interativo
if __name__ == "__main__":
    menu()
```
